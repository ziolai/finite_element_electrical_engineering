```python
import numpy as np
import sympy as sym
from sympy import symbols, Function, sqrt, integrate, diff, asinh, simplify
```

```python
s   = symbols("s", real=True)
h   = symbols("h", positive=True)
eta = symbols("eta", real=True) 
R0  = symbols("R0", positive=True)
f1  = Function('f1', real=True)
I8integrand  = Function('I8integrand')
```

```python
f1 = (sqrt(h**2+eta**2)-h) / eta**2 
I8integrand = f1.evalf(subs={eta:sqrt(R0**2+s**2)})
```

```python
diff(asinh(s/sqrt(h**2+R0**2)),s)
```

```python
simplify(integrate(I8integrand,(s,0,1)))
```
