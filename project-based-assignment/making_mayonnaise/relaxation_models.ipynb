# ODEs Models for Relaxations

Inspired by [Relaxation and Rheology near Jamming](https://arxiv.org/pdf/1106.0714). Trying to understand the appearance of a plateau region in the function $G'(\omega)$.  

Considering a [Kelvin Voigt material](https://en.wikipedia.org/wiki/Kelvin–Voigt_material) with spring constant $E$ (physical dimensions?) and damper constant $\eta$ (physical dimensions?) an for which the stress $\sigma(t)$ - strain $\epsilon(t)$ relation is given by 

$$
\sigma(t) = E \epsilon(t) + \eta \dot{\epsilon}(t) \, . 
$$

The corresponding relaxation time $\tau_R$ is given $\tau_R = \frac{\eta}{E}$. (physical dimensions of time?) 

<b>1DOF Model Without External Forcing</b>: What follows is true for a [Kelvin-Voigt material](https://en.wikipedia.org/wiki/Kelvin–Voigt_material). Allow strain to evolve from initial value. Problem to solve  

$$
\eta \dot{\epsilon}(t) + E \epsilon(t) = 0 \text{ with initial condition } \epsilon(t=0) = \epsilon_0 \, . 
$$

or

$$
\dot{\epsilon}(t) + \frac{1}{\tau_R} \epsilon(t) = 0 \text{ with initial condition } \epsilon(t=0) = \epsilon_0 \, . 
$$

has analytical solution 

$$
\epsilon(t) =  \epsilon_0 \exp(- t/ \tau_R) . 
$$

Exponential decrease with characteristic time constant $\tau_R$. Small value of $\tau_R$ implies fast transition to unstressed situation. How to adapt this material for a [Maxwell material](https://en.wikipedia.org/wiki/Kelvin–Voigt_material)?  

<b>1DOF Model With Sinusoidal Forcing</b>: Problem to Solve

$$
\dot{\epsilon}(t) + C \epsilon(t) = F_d \sin(\omega_0 \, t) \text{ given } \epsilon(t=0) =  \epsilon_0 \, . 
$$

has analytical solution (using integrating factor $M(t) = \exp(C t)$)

$$
\epsilon(t) = \epsilon_0 \exp(- t/ \tau_R) + F_d \frac{\tau_R}{1+\tau_R^2 \omega_0^2} 
              [ \sin(\omega_0 t) - \tau_R \omega_0 \cos(\omega_0 t)]. 
$$

First term is a vanishing transient solution. Second term is a scaled and phased shift of the driving force. We have that 
1. if $\tau_R^2 \omega_0^2 \ll 1$ or equivalently if $\omega_0 \ll 1/\tau_R$ (low frequency limit), then the scaled amplitude equals $F_d \tau_r$;
2. if $\tau_R^2 \omega_0^2 = 1$, then the scaled amplitude equals $F_d \tau_r/2$
3. if $\tau_R^2 \omega_0^2 \gg 1$ or equivalently if $\omega_0 \gg 1/\tau_R$ (high frequency limit), then the scaled amplitude is zero;
