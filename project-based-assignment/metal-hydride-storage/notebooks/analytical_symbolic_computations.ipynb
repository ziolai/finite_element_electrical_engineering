## Section 1: Steady State Convection - Diffusion Equation 

Let $x \in \Omega = (0,1)$ denote the spatial domain. Let $u(x)$ denote the concentration of a dye (or temperature, or concentration of a gas) that diffuses and is convected on $\Omega$. We wish to find the steady state concentration. We thus solve the boundary value problem for the stationary [convection-diffusion](https://en.wikipedia.org/wiki/Convection–diffusion_equation) equation in one spatial dimension for the unknown $u(x)$ 

$$
u'(x) = \epsilon \, u''(x) + f(x) 
$$

subject to the Dirichlet boundary conditions $u(0)=1$ and $u(1)=0$. The number $\epsilon$ is the [Peclet number](https://en.wikipedia.org/wiki/Péclet_number). 

```python
from sympy import Function, dsolve, Derivative, plot
from sympy.abc import x

u = symbols("u", cls=Function)
f = symbols("f", cls=Function)
x = symbols("x")
eps = symbols("eps", positive=True)

f = x 
# Solve the ODE
eqns = eps*Derivative(u(x), x, x) - Derivative(u(x), x) + f 
ics0 = {u(0): 1, u(1): 0}
sol = dsolve(eqns,ics = ics0)

plotb = plot(show=False)
for epsi in sym.Array([1.,0.1,0.01]):
    plotb1 = plot(sol.rhs.subs(eps,epsi), (x, 0, 1), show=False)
    plotb.append(plotb1[0])
plotb.show()
```
